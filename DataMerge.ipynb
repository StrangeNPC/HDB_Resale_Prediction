```python
import pandas as pd
import datetime
```

```python
HDB_Resale_data=pd.read_excel('HDBWithLatLong.xlsx',index_col=0)
```

```python
df=HDB_Resale_data
```

```python
# Convert the "Months" column to years
df["Years"] = df["month"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m").year)

# Drop the original "Months" column if needed
df.drop(columns=["month"], inplace=True)

# Get the column names in the current order
columns = df.columns.tolist()

# Move the "resale_price" column to the extreme right
columns.remove("resale_price")
columns.append("resale_price")

# Reorder the columns in the DataFrame
df = df[columns]

#remove months 2023

# Filter out rows with year == 2023
df = df[df.Years != 2023]
```

```python
def objective(trial):
    param = {
        'max_depth': trial.suggest_int('max_depth', 1, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 1.0),
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_float('gamma', 0.01, 1.0),
        'subsample': trial.suggest_float('subsample', 0.01, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.01, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.01, 1.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.01, 1.0),
        'random_state': trial.suggest_int('random_state', 1, 1000)
    }
    model = xgboost.XGBRegressor(**param)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return mean_squared_error(y_test, y_pred)
```

```python
# Create the study
study = optuna.create_study(direction='minimize', study_name='regression')
study.optimize(objective, n_trials=100)
```
